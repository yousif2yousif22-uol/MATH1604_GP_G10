## Walkthrough of my Functions

### `setup_folders()`

### `step2_extract_seq()`
Step 2 loops through the answers files  in the data folder and uses M1's module to process each one of them. For each file, it calls `extract_answers_sequence()` to read the raw quiz tiles and return a list of 100 integers. The answers correspond to values 1 to 4, with unanswered questions being 0. Then, it calls `write_answers_sequence()` to pass the extracted list, ID, and output folder path and saves this info to files (warning and skipping any non-existent files).

### `step3_analyse_visualise()`
The third step is to use M3's module to analyze the collated data and create visualizations. It starts by calling `generate_means_sequence()` to read the collated files, split them into singular respondent sections, extracts each sequence and calculates the mean answer for each of the 100 questions (not including unanswered).This results in a list of 100 floats. It then uses `visualize_data()` to create the scatter plot of the means and the line plot with the individual answer sequences.

## Analysis of the Results
- The scatter plot shows that the mean answer to all the questions are tightly clustered around each of the 4 number values (either 1,2,3 or 4) with mostly no means near intermediate values like 2.5 or 3.5.
If the respondents were guessing randomly, we would have means around 2.5 for every question. However the fact that each questions mean is toward a specific whole number shows that the quiz consists of a deliberate correct answer sequence, with the majority of the respondents answering correctly. 

- The line plot shows all respondents answers on the same plot, which looks messy at first, with lines going between 1 and 4. However, this supports the scatter plot, even though the answers vary, their averages still group up around specific values for each question, showing that consistent correct answers are pulling the means toward it.
