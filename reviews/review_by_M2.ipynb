## Testing the collated output file

After downloading the files, I tested the collation function.

The function created:

`output/collated_answers.txt`

I checked that the file exists. I also counted the number of separator lines. Since there are 64 respondent files, there should be 63 separator lines between them.


```python
collated_path = "output/collated_answers.txt"

print("Collated file exists:", os.path.exists(collated_path))

separator_count = 0

with open(collated_path, "r", encoding="utf-8") as file:
    for line in file:
        if line.strip() == "*":
            separator_count += 1

print("Number of separator lines:", separator_count)
```
