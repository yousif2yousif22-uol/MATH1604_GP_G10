## Overview
This notebook is a review of the data analysis module I created, which is responsible for processing and visualising the quiz response data. The ojbective of this module is to identify potential patterns in the sequence of correct answers by aggregating responses and analysing their distribution.

I created two functions: `generate_means_sequence` which computes the mean answer for each question and `visualize_data` which outputs either a scatter plot, or a line plot which can be used to investigate whether the answer distribution exhibits any structured behaviour.

## Walkthrough of my functions

### `generate_means_sequence`

This function starts by opening the `collated_answers.txt` and reading its contents as a single string. It then splits the file into seperate respondent blocks, ensuring each respondent's data can be processed independently.
 Each respondent is temporarily written to a file and passed to the `extract_answers_sequence` function from the M1 module. This step converts the raw data into a list of numbers, representing the selected answer. Once all the respondents have been extracted, the function iterates over each of the questions and it calculates their mean value (excluding zeroes). This aggregation allows patterns to be identified that are not immediately visible in individual response data.
 Finally, the function returns a list of 100 floating-point values, where each value is the mean answer for a specific question.

### `visualize_data`

This function uses the previous function to create graphical representations of the dataset to help analyse the data.
It starts by splitting the `collated_answers.txt` again and then parsed the answers through the `extract_answers_sequence` function. It then creates the x-axis representing question numbers from 1 to 100. Based on the value of the parameter `n`, the function generates different outputs:
- When `n = 1`, the function uses `generate_means_sequence` to compute the mean answer values and display them using a scatter plot
- When `n = 2`, the function plots each respondent's answer sequence as a line on the same graph, allowing for comparison between individuals
- if `n` is neither 1 or 2, the function outputs an error message

## Analysis of Results

The scatter plot shows that the answers are tightly clustered around the number values. This indicates that, for most questions, the average response tends to be strongly influenced by one dominant choice. This shows that people were consistently selecting similar options for each question, and were unlikely to be guesses.
Also, the relative stability of the mean values suggest there is no obvious upward or downward trend throughout the quiz. Instead, each question exhibits its own characteristic mean level but seems to be generally stay quite stationary. This implies that the sequence may follow a structured but non-trending pattern.

The line plot gives more insights as the plot shows a lot of variability between respondents, with a lot of changes between the answer values, and occasional zeros indicating when people didn't answer a question. However, the density of overlapping lines suggests the response are not distributed amongst all values, but usually more concentrated at one value for different questions, which helps show the clustering seen in the mean plot. We can see how a majority always point to one value, and this plot is helpful in seeing the people who didn't get it correct or skipped the question.

## Limitations
There are useful insights that can be taken from this, but there are several limitations that come up.
- The use of the mean as a summary statistic doesn't capture the full distribution of responses. Differen answer patterns can produce the same mean value and so some important insights may be lost. Such as, a question with two dominant answer choices may have the same mean as one with just one answer.
- Any patterns in the ordering of answers aren't clear. This method only looks at each question on its own and calculates a mean, and doesn't consider how answers might relate to each other (such as guessing in a consistent way).
- The line plot looks quite messy and the noise makes it hard to identify patterns at the individual level
