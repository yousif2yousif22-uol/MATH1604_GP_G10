## Testing the collated output file

After downloading the respondent files, I used my collation function to create `output/collated_answers.txt`.

I then checked whether the collated file exists and counted the number of separator lines. Since there are 64 respondent files, there should be 63 separator lines between them.


```python
collated_path = "output/collated_answers.txt"

print("Collated file exists:", os.path.exists(collated_path))

separator_count = 0

with open(collated_path, "r", encoding="utf-8") as file:
    for line in file:
        if line.strip() == "*":
            separator_count += 1

print("Number of separator lines:", separator_count)
```
